```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_main=pd.read_csv('https://raw.githubusercontent.com/D3bug1t/ExoStellar_Nexus/6eead614ef932da8a2421f13bb22ccad44f45d78/Assignments/Assignment-2/advertising.csv')
 
```

```python
dataset_main
```

```python
y=dataset_main['Sales']
x1=dataset_main['TV']
x2=dataset_main['Radio']
x3=dataset_main['Newspaper']
```

```python
plt.scatter(x1,y)
plt.xlabel('TV') 
plt.ylabel('Sales')
```

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y, test_size=0.4, random_state=32)
x1_train=np.array(x1_train).reshape(-1,1)
x1_test=np.array(x1_test).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr1=LinearRegression()
lr1.fit(x1_train,y1_train)
```

```python
c1=lr1.intercept_
c1
```

```python
m1=lr1.coef_
m1
```

```python
y1_train_pred=lr1.predict(x1_train)
plt.plot(x1_train,y1_train_pred, color="red")
plt.scatter(x1_train,y1_train)
```

```python
y1_test_pred=lr1.predict(x1_test)
plt.plot(x1_test,y1_test_pred, color="red")
plt.scatter(x1_test,y1_test)
```

```python

y1_test=np.array(y1_test)
y1_avg=0
rss1=0
tss1=0
for i in range(80):
    y1_avg=y1_avg+y1_test[i]
    s1=(y1_test[i]-y1_test_pred[i])*(y1_test[i]-y1_test_pred[i])
    rss1=s1+rss1

for i in range(80):
    s_1=(y1_test[i]-y1_avg)*(y1_test[i]-y1_avg)
    tss1=tss1+s_1    
    
r2_1=1-(rss1/tss1)
r2_1
```
